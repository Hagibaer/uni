```python
import pandas as pd
import os
import numpy as np
from collections import Counter
```

```python
for file in os.listdir('./corpus/email-korpus/train-ham'):
     if os.path.isfile(file):
        print (file)
```

```python
emails = [os.path.join('./corpus/email-korpus/train-ham',f) for f in os.listdir('./corpus/email-korpus/train-ham')]
in_body = False
words_in_body = []
for mail in emails:
    with open(mail) as m:
        for line in m:
            if not in_body:
                if line == '\n':
                    in_body = True
            else:
                words = line.split()
                words_in_body += words

word_dict = Counter(words_in_body)
list_to_remove = list(word_dict)

for item in list_to_remove:
    if item.isalpha() == False:
        del word_dict[item]
    elif len(item) == 1:
        del word_dict[item]

word_dict = word_dict.most_common(5000)

word_dict
```
